## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']
print(df.head(1))
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['shipping_price'] = np.where(df['unit_weight'] * df['qty'] > 50,
                               7 * df['unit_weight'] * df['qty'],
                               10 * df['unit_weight'] * df['qty'])
print(df.head())
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df['total_price'] = 1.0925 * (df['subtotal'] + df['shipping_price'])
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
print(df.head())
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = df['total_price'] - df['line_cost']
print(df.head())
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
client_spending = {}

for client_id in top_five_clients_ids:
    client_orders = df[df['client_id'] == client_id]
    total_spent = client_orders['total_price'].sum()
    client_spending[client_id] = total_spent
    
for client_id, spent in client_spending.items():
    print(f"Client ID {client_id} spent a total of ${spent:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.


# Filter the DataFrame for the top five clients
top_clients_df = df[df['client_id'].isin(top_five_clients_ids)]

# Group by 'client_id' and aggregate the straightforward data
summary_df = top_clients_df.groupby('client_id').agg(
    Total_Units=('qty', 'sum'),
    Total_Shipping_Price=('shipping_price', 'sum'),
    Total_Revenue=('total_price', 'sum')
).reset_index()

# Calculate Total Profit separately
total_profit = top_clients_df.groupby('client_id').apply(
    lambda x: (x['total_price'] - x['line_cost']).sum()
)
summary_df['Total_Profit'] = total_profit.values

# Sort the DataFrame by Total Profit in descending order
summary_df.sort_values(by='Total_Profit', ascending=False, inplace=True)

# Display the summary DataFrame
print(summary_df)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Convert monetary values to millions
summary_df['Total_Units'] = summary_df['Total_Units']  # Keep as is, not a monetary value
summary_df['Total_Shipping_Price'] = summary_df['Total_Shipping_Price'] / 1_000_000  # Convert to millions
summary_df['Total_Revenue'] = summary_df['Total_Revenue'] / 1_000_000  # Convert to millions
summary_df['Total_Profit'] = summary_df['Total_Profit'] / 1_000_000  # Convert to millions

# Optionally, format numbers to two decimal places
summary_df['Total_Shipping_Price'] = summary_df['Total_Shipping_Price'].round(2)
summary_df['Total_Revenue'] = summary_df['Total_Revenue'].round(2)
summary_df['Total_Profit'] = summary_df['Total_Profit'].round(2)

# Rename columns for better presentation
summary_df.rename(columns={
    'Client ID': 'Client ID',
    'Total_Units': 'Total Units Purchased',
    'Total_Shipping_Price': 'Total Shipping Cost (Million $)',
    'Total_Revenue': 'Total Revenue (Million $)',
    'Total_Profit': 'Total Profit (Million $)'
}, inplace=True)

# Display the formatted summary DataFrame
print(summary_df)
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df.sort_values(by='Total Profit (Million $)', ascending=False, inplace=True)

# Display the sorted DataFrame
print(summary_df)
```
